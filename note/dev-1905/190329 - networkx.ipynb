Use this to create super-easy plots of chains and ensembles. First move towards autoML/ILP.

This is really something very interesting to think about. **The idea should indeed be a kind of `SQL`, but then not for databases, but for data and actions on it.**

Category theory is the perfect tool to formulate this. Diagrams can help, and if you then can specify a sufficiently general language for this kind of thing, it should be something pretty nice.

Especially when recursion comes into play. And tensors.

```python
import networkx as nx
import nxviz as nxv
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(['d1', 'd2'])
G.add_nodes_from(['m1'])
```

```python
G.add_nodes_from(['out'])
```

```python
list(G.nodes())
```

```python
G.add_edges_from([('d1', 'm1', {'id':1}), ('d2', 'm1', {'id':2}), ('m1', 'out', {'id':3})])
```

```python
nx.draw(G, with_labels=True)
```

```python
list(G.edges(data=True))
```

```python
list(G.nodes(data=True))
```

## Now for real

```python
B = nx.DiGraph()

# Add nodes with the node attribute "bipartite"
B.add_nodes_from(['d0', 'd1', 'd3'], bipartite=0)
B.add_nodes_from(['m0', 'm1'], bipartite=1)

# Add edges only between nodes of opposite node sets
B.add_edges_from([('d0','m0'), ('d1','m0'), ('m0', 'd3'),
                  ('d1','m1'), ('d3','m1'), ('m1', 'd0')])
```

```python
nx.draw(B, with_labels=True)
```

```python
from networkx.algorithms import bipartite
bottom_nodes, top_nodes = bipartite.sets(B)
```

```python
nx.draw(B, with_labels=True)
```
